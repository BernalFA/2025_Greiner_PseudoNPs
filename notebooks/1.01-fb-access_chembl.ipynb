```python
sql = """
SELECT DISTINCT md.chembl_id AS Chembl_Id,
                cs.canonical_smiles AS Canonical_Smiles

FROM molecule_dictionary md
JOIN compound_structures cs ON cs.molregno = md.molregno

WHERE md.natural_product IN ('1')
"""
```

```python
cur.execute(sql)
res = cur.fetchall()
conn.close()
print(len(res))
```
